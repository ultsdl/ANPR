```python
import xml.etree.ElementTree as ET
from os import getcwd
import os
from tqdm.notebook import tqdm
import shutil
```

```python
dataset_train = os.path.join(os.getcwd(),'train_test\\allfiles')
```

```python
wd = getcwd()
```

```python
all_files_path = os.listdir(dataset_train)
all_files = [dataset_train+"\\"+i for i in all_files_path]
```

```python
len(all_files)
```

```python
rejected_list = []
for fullname in tqdm(all_files):
    if not fullname.endswith('.jpg'):
        continue
    if not os.path.exists(fullname[:-4]+".xml"):
        rejected_list.append(fullname)
        shutil.copy2(fullname, os.getcwd()+"/images")
```
